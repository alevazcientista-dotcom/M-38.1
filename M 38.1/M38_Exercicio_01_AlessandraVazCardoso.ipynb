## 1. Carregamento e InspeûÏûÈo dos Dados

Utilizamos o dataset `tips`, que contûˋm registros de gorjetas em um restaurante ao longo de vûÀrios dias. As variûÀveis disponûÙveis sûÈo:

| VariûÀvel      | Tipo         | DescriûÏûÈo                                  |
|--------------|-------------|--------------------------------------------|
| `total_bill` | Numûˋrica     | Valor total da conta (USD)                 |
| `tip`        | Numûˋrica     | Valor da gorjeta (USD)                     |
| `sex`        | Categû°rica   | Sexo do pagante (Male / Female)            |
| `smoker`     | Categû°rica   | Fumante? (Yes / No)                        |
| `day`        | Categû°rica   | Dia da semana (Thur / Fri / Sat / Sun)     |
| `time`       | Categû°rica   | PerûÙodo (Lunch / Dinner)                   |
| `size`       | Numûˋrica     | Nû¤mero de pessoas na mesa                  |

```python
# Carregando o dataset
df = sns.load_dataset('tips')

print(f'Shape do dataset: {df.shape[0]} linhas û {df.shape[1]} colunas')
print('\nPrimeiras 5 linhas:')
df.head()
```

```python
# InformaûÏûçes gerais sobre tipos e missings
print('=== InformaûÏûçes do Dataset ===')
df.info()
print('\nValores ausentes por coluna:')
print(df.isnull().sum())
```

```python
# EstatûÙsticas descritivas gerais
print('=== EstatûÙsticas Descritivas ã VariûÀveis Numûˋricas ===')
df.describe().round(2)
```

## 2. Tabelas de Frequûˆncia

As **tabelas de frequûˆncia** mostram quantas vezes cada categoria aparece nos dados, tanto em termos absolutos quanto relativos (percentual).

```python
# FunûÏûÈo auxiliar para tabela de frequûˆncia formatada
def tabela_frequencia(serie, nome_variavel):
    freq_abs  = serie.value_counts()
    freq_rel  = serie.value_counts(normalize=True).mul(100).round(2)
    freq_acum = freq_rel.cumsum().round(2)

    tabela = pd.DataFrame({
        'Frequûˆncia Absoluta': freq_abs,
        'Frequûˆncia Relativa (%)': freq_rel,
        'Frequûˆncia Acumulada (%)': freq_acum
    })
    tabela.index.name = nome_variavel
    return tabela

# VariûÀveis categû°ricas
variaveis_cat = ['sex', 'smoker', 'day', 'time']

for var in variaveis_cat:
    print(f'\n{'='*55}')
    print(f'  Tabela de Frequûˆncia ã {var.upper()}')
    print(f'{'='*55}')
    display(tabela_frequencia(df[var], var))
```

## 3. Tabelas Cruzadas (Crosstab)

As **tabelas cruzadas** (ou tabelas de contingûˆncia) mostram a distribuiûÏûÈo conjunta de duas variûÀveis categû°ricas, revelando possûÙveis relaûÏûçes entre elas.

```python
# Tabela cruzada: sexo û dia da semana
print('=== Tabela Cruzada: Sexo û Dia da Semana (contagem) ===')
ct1 = pd.crosstab(df['sex'], df['day'], margins=True, margins_name='Total')
display(ct1)

print('\n=== Tabela Cruzada: Sexo û Dia da Semana (proporûÏûÈo por linha, %) ===')
ct1_pct = pd.crosstab(df['sex'], df['day'], normalize='index').mul(100).round(2)
display(ct1_pct)
```

```python
# Tabela cruzada: fumante û perûÙodo (Lunch/Dinner)
print('=== Tabela Cruzada: Fumante û PerûÙodo (contagem) ===')
ct2 = pd.crosstab(df['smoker'], df['time'], margins=True, margins_name='Total')
display(ct2)

print('\n=== Tabela Cruzada: Fumante û PerûÙodo (proporûÏûÈo por coluna, %) ===')
ct2_pct = pd.crosstab(df['smoker'], df['time'], normalize='columns').mul(100).round(2)
display(ct2_pct)
```

```python
# Tabela cruzada com valor mûˋdio (gorjeta mûˋdia por sexo e dia)
print('=== Gorjeta Mûˋdia (USD) por Sexo e Dia da Semana ===')
pivot_tip = df.pivot_table(
    values='tip',
    index='sex',
    columns='day',
    aggfunc='mean'
).round(2)
display(pivot_tip)
```

## 4. EstatûÙsticas Descritivas por Grupo

Aqui calculamos medidas de tendûˆncia central e dispersûÈo segmentadas por variûÀveis categû°ricas de interesse.

```python
# Descritivas de total_bill e tip agrupadas por dia
print('=== EstatûÙsticas de total_bill e tip por Dia da Semana ===')
desc_dia = df.groupby('day')[['total_bill', 'tip']].agg(
    ['mean', 'median', 'std', 'min', 'max', 'count']
).round(2)
display(desc_dia)
```

```python
# Descritivas segmentadas por sexo e perûÙodo
print('=== Gorjeta Mûˋdia e Mediana por Sexo e PerûÙodo ===')
desc_grp = df.groupby(['sex', 'time'])['tip'].agg(
    media='mean',
    mediana='median',
    desvio_padrao='std',
    n='count'
).round(2)
display(desc_grp)
```

```python
# Criando variûÀvel: percentual de gorjeta sobre a conta total
df['pct_gorjeta'] = (df['tip'] / df['total_bill'] * 100).round(2)

print('=== Percentual de Gorjeta sobre a Conta (%) ===')
print(df['pct_gorjeta'].describe().round(2))
print(f'\nPercentual mûˋdio de gorjeta: {df["pct_gorjeta"].mean():.2f}%')
print(f'Percentual mediano de gorjeta: {df["pct_gorjeta"].median():.2f}%')
```

## 5. VisualizaûÏûçes GrûÀficas

### 5.4 Boxplot ã ComparaûÏûÈo da Gorjeta por Dia da Semana

```python
ordem_dias = ['Thur', 'Fri', 'Sat', 'Sun']
paleta     = ['#4C72B0', '#55A868', '#C44E52', '#8172B2']

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Boxplot: gorjeta por dia
sns.boxplot(
    data=df, x='day', y='tip',
    order=ordem_dias, palette=paleta,
    ax=axes[0], linewidth=1.2
)
axes[0].set_title('Gorjeta por Dia da Semana')
axes[0].set_xlabel('Dia da Semana')
axes[0].set_ylabel('Gorjeta (USD)')

# Boxplot: valor da conta por perûÙodo
sns.boxplot(
    data=df, x='time', y='total_bill',
    palette=['#DD8452', '#4C72B0'],
    ax=axes[1], linewidth=1.2
)
axes[1].set_title('Valor da Conta por PerûÙodo')
axes[1].set_xlabel('PerûÙodo')
axes[1].set_ylabel('Valor da Conta (USD)')

fig.suptitle('Boxplots ã AnûÀlise por Grupos', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

### 5.5 GrûÀfico de DispersûÈo ã Conta Total û Gorjeta

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# DispersûÈo simples com linha de tendûˆncia
axes[0].scatter(
    df['total_bill'], df['tip'],
    alpha=0.6, color='#4C72B0', edgecolors='white', linewidths=0.4, s=60
)
# Linha de tendûˆncia
m, b = np.polyfit(df['total_bill'], df['tip'], 1)
x_line = np.linspace(df['total_bill'].min(), df['total_bill'].max(), 100)
axes[0].plot(x_line, m * x_line + b, color='red', linewidth=2, label=f'y = {m:.3f}x + {b:.3f}')
axes[0].set_title('Conta Total û Gorjeta')
axes[0].set_xlabel('Valor da Conta (USD)')
axes[0].set_ylabel('Gorjeta (USD)')
axes[0].legend(fontsize=9)

# DispersûÈo colorida por sexo
cores = {'Male': '#4C72B0', 'Female': '#DD8452'}
for sexo, grupo in df.groupby('sex'):
    axes[1].scatter(
        grupo['total_bill'], grupo['tip'],
        alpha=0.65, color=cores[sexo],
        edgecolors='white', linewidths=0.4,
        s=60, label=sexo
    )
axes[1].set_title('Conta Total û Gorjeta (por Sexo)')
axes[1].set_xlabel('Valor da Conta (USD)')
axes[1].set_ylabel('Gorjeta (USD)')
axes[1].legend(title='Sexo')

# CorrelaûÏûÈo
corr = df['total_bill'].corr(df['tip'])
fig.suptitle(f'DispersûÈo: Conta Total û Gorjeta  |  CorrelaûÏûÈo de Pearson: {corr:.3f}',
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

print(f'\nCorrelaûÏûÈo de Pearson (total_bill û tip): {corr:.4f}')
```

## 6. AnûÀlise do Percentual de Gorjeta

O **percentual de gorjeta** (`pct_gorjeta`) representa a proporûÏûÈo da gorjeta em relaûÏûÈo ao valor total da conta. Essa mûˋtrica permite comparar o comportamento de gorjeta de forma normalizada, independentemente do valor da conta.

```python
# DistribuiûÏûÈo do percentual de gorjeta por sexo e fumante
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

sns.violinplot(
    data=df, x='sex', y='pct_gorjeta',
    palette=['#4C72B0', '#DD8452'],
    inner='quartile', linewidth=1.3,
    ax=axes[0]
)
axes[0].set_title('Percentual de Gorjeta por Sexo')
axes[0].set_xlabel('Sexo')
axes[0].set_ylabel('% Gorjeta sobre a Conta')

sns.violinplot(
    data=df, x='smoker', y='pct_gorjeta',
    palette=['#55A868', '#C44E52'],
    inner='quartile', linewidth=1.3,
    ax=axes[1]
)
axes[1].set_title('Percentual de Gorjeta por HûÀbito de Fumar')
axes[1].set_xlabel('Fumante')
axes[1].set_ylabel('% Gorjeta sobre a Conta')

fig.suptitle('DistribuiûÏûÈo do Percentual de Gorjeta (Violin Plot)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Tabela resumo do percentual de gorjeta por grupo
print('=== Percentual Mûˋdio de Gorjeta por Grupo ===')
resumo_pct = df.groupby(['sex', 'smoker'])['pct_gorjeta'].agg(
    media='mean',
    mediana='median',
    desvio_padrao='std',
    n='count'
).round(2)
display(resumo_pct)
```
